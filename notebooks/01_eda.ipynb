```python
plt.figure(figsize=(8, 5))
plt.hist(train['SalePrice'], bins=50, edgecolor='black')
plt.title('SalePrice Distribution')
plt.xlabel('SalePrice')
plt.ylabel('Count')
plt.show()

print(f"Среднее: {train['SalePrice'].mean():,.0f}")
print(f"Медиана: {train['SalePrice'].median():,.0f}")
print(f"Skewness: {train['SalePrice'].skew():.2f}")
```

```python
fig, axes = plt.subplots(1, 2, figsize=(14, 5))

axes[0].hist(train['SalePrice'], bins=50, edgecolor='black')
axes[0].set_title(f"SalePrice (skew={train['SalePrice'].skew():.2f})")
axes[0].set_xlabel('SalePrice')

axes[1].hist(np.log1p(train['SalePrice']), bins=50, edgecolor='black')
axes[1].set_title(f"log(SalePrice) (skew={np.log1p(train['SalePrice']).skew():.2f})")
axes[1].set_xlabel('log(SalePrice)')

plt.tight_layout()
plt.show()
```

```python
correlation = train.corr(numeric_only=True)['SalePrice'].drop('SalePrice')
top_corr = correlation.abs().sort_values(ascending=False).head(10)

print("Топ-10 фичей по корреляции с SalePrice:")
print()
for feat in top_corr.index:
    print(f"  {feat:20s} {correlation[feat]:+.3f}")
```

```python
top_features = ['OverallQual', 'GrLivArea', 'GarageCars', 'GarageArea',
                'TotalBsmtSF', '1stFlrSF', 'FullBath', 'TotRmsAbvGrd',
                'YearBuilt', 'YearRemodAdd']

fig, axes = plt.subplots(2, 5, figsize=(20, 8))
axes = axes.flatten()

for i, feat in enumerate(top_features):
    axes[i].scatter(train[feat], train['SalePrice'], alpha=0.3, s=10)
    axes[i].set_xlabel(feat)
    axes[i].set_ylabel('SalePrice')
    axes[i].set_title(f'{feat} (corr={correlation[feat]:+.2f})')

plt.tight_layout()
plt.show()
```

```python
top_features_with_target = top_features + ['SalePrice']
corr_matrix = train[top_features_with_target].corr()

plt.figure(figsize=(12, 8))
sns.heatmap(corr_matrix, annot=True, fmt='.2f')
plt.tight_layout()
plt.show()
```

## EDA Summary

Train: 1460 строк, 81 колонка (38 числовых, 43 категориальных). Test: 1459 строк, 80 колонок.

**Target (SalePrice):** 
диапазон от 34,900 до 755,000. Среднее 180,921, медиана 163,000 (среднее > медианы из-за дорогих домов в хвосте). Распределение скошено вправо (skewness=1.88), log-трансформация выравнивает до 0.12.

**Пропуски:** 
19 колонок. Большинство — не настоящие пропуски, а отсутствие объекта (нет гаража, подвала, бассейна и т.д.). 
Реальные пропуски: LotFrontage (259), MasVnrType/MasVnrArea (8), Electrical (1).

**Почти пустые фичи:** 
PoolArea, 3SsnPorch, ScreenPorch, LowQualFinSF, BsmtFinSF2, MiscVal — у 75%+ домов значение = 0. Скорее всего малополезны для модели.

**Топ фичи по корреляции с ценой:** 
OverallQual (0.79), GrLivArea (0.71), GarageCars (0.64), GarageArea (0.62), TotalBsmtSF (0.61), 1stFlrSF (0.61).

**Скоррелированные пары фичей (0.8+):** 
GarageCars—GarageArea (0.88), GrLivArea—TotRmsAbvGrd (0.83), TotalBsmtSF—1stFlrSF (0.82).

**Выбросы:** 
GrLivArea — 2 дома >4000 кв.футов с ценой <200,000. TotalBsmtSF — точки >5000. LotArea — max 215,245 при среднем 10,516.

**Категориальные фичи:** 
ExterQual, KitchenQual, BsmtQual хорошо разделяют цену. Neighborhood — сильное влияние, но 25 категорий.
